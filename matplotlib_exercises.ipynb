2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

        y =√x
        
        y = x**3
        
        y = 2**x
        
        y = 1/(x+1)
        
You can use functions from the math module to help implement some of the equations above.

```python
x = [1,2,3,4,5]

df_one_data = {'x':x,
               'y':[math.sqrt(n) for n in x]}
df_one = pd.DataFrame(df_one_data)

df_two_data = {'x':x,
               'y':[n**3 for n in x]}
df_two = pd.DataFrame(df_two_data)

df_three_data = {'x':x,
               'y':[2**n for n in x]}
df_three = pd.DataFrame(df_three_data)

df_four_data = {'x':x,
               'y':[(1/(n+1)) for n in x]}
df_four = pd.DataFrame(df_four_data)
```

```python
plt.plot(df_one.x,df_one.y)
```

```python
plt.plot(df_two.x,df_two.y)
```

```python
plt.plot(df_three.x,df_three.y)
```

```python
plt.plot(df_four.x,df_four.y)
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize=(10,10))

#subplot 1
plt.subplot(2,2,1)
plt.plot(df_one.x,df_one.y)

#subplot 2
plt.subplot(2,2,2)
plt.plot(df_two.x,df_two.y)

#subplot 3
plt.subplot(2,2,3)
plt.plot(df_three.x,df_three.y)

#subplot 4
plt.subplot(2,2,4)
plt.plot(df_four.x,df_four.y)
```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.figure(figsize=(8,8))

plt.plot(df_one.x,df_one.y,label='y =√x')

plt.plot(df_two.x,df_two.y,label='y = x**3')

plt.plot(df_three.x,df_three.y,label='y = 2**x')

plt.plot(df_four.x,df_four.y,label='y = 1/(x+1)')

plt.title('Comparison of different equations')

plt.legend()
```
